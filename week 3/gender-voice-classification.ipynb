```python
voice_data = pd.read_csv('/kaggle/input/voicegender/voice.csv')
```

```python
voice_data.isnull().sum() 
```

```python

x = voice_data.iloc[:, :-1].values
y = voice_data.iloc[:,-1].values
```

```python
x
```

```python

y
```

```python
gender_encoder = LabelEncoder()
y = gender_encoder.fit_transform(y) #male = 1, female = 0 (i think)
y
```

```python

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .2)
```

```python


column = x_train.shape[1]
y_train = to_categorical(y_train, 2)

hist = History()

model = Sequential()


model.add(Dense(1000, activation='relu', ))
model.add(Dense(1000, activation='relu'))
model.add(Dense(1000, activation='relu'))
model.add(Dense(1000, activation='sigmoid'))


model.add(Dense(2, activation='softmax'))

model.compile(optimizer='adam', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs = 40, validation_split = .1, callbacks = [hist])
```

```python
y_pred =  model.predict(x_test)
y_pred = np.round(y_pred[:,1])
print(metrics.accuracy_score(y_pred,y_test))
```
